```python
import pandas as mypd
from sklearn import tree
from sklearn.cross_validation import train_test_split
```

```python
mydata = mypd.read_csv("/Users/harisuriya/Desktop/mlworkshop/Data/bank-data.csv")
```

```python
mydata
```

```python
x = mydata.values[:, 0:9]
y = mydata.values[:, 10]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
mymodel = tree.DecisionTreeClassifier(min_samples_split=50)
mymodel.fit(x_train, y_train)

```

```python
mymodel.score(x_train, y_train)
```

```python
pred = mymodel.predict(x_train)
```

```python
mytable = mypd.crosstab(y_train, pred)
```

```python
mytable
```

```python
pred = mymodel.predict(x_test)
```

```python
mytable = mypd.crosstab(y_test, pred)
```
